```python

filename = "fengshui.txt"
with open(filename) as f:
 mytext = f.read()
import jieba
mytext = " ".join(jieba.cut(mytext))
from wordcloud import WordCloud,STOPWORDS

stopwords = set(STOPWORDS)
stopwords.add(",,")
wordcloud = WordCloud(background_color="white",width=1000, height=860, margin=2,font_path="simsun.ttf",stopwords=stopwords).generate(mytext)
%pylab inline
import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.savefig('pink.png', dpi=300)
```

```python
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

text = open(path.join( 'fengshui.txt')).read()
alice_coloring = np.array(Image.open(path.join( "bagua.jpg")))

stopwords = set(STOPWORDS)
stopwords.add(",,")
wc = WordCloud(font_path="simsun.ttf",background_color="white", max_words=2000, mask=alice_coloring,width=1000, height=860,
               stopwords=stopwords, max_font_size=80, random_state=42)
wc.generate(text)
image_colors = ImageColorGenerator(alice_coloring)

plt.figure()
plt.imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")
plt.figure()

```
