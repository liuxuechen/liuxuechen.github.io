```python
%%time
import requests
import csv
import pandas as pd
from bs4 import BeautifulSoup
from time import sleep
from random import randint
from time import time

header = ['page #','title','url','details','post date']
data = []

for i in range(1,10):
 quote_page = requests.get('https://www.discuss.com.hk/forumdisplay.php?fid=204&page='+str(i))
 sleep(randint(8,15))
 print("\n***** Page " + str(i) +" in action *****")
 soup = BeautifulSoup(quote_page.content,'html.parser')

 for article in soup.find_all("span", attrs={"class":"tsubject"}):

  page_no = str(i)
  title = article.a.text.encode('utf-8').strip()
  decoded_title = title.decode('utf-8')
  print(decoded_title)
  data.append((decoded_title))

with open('fengshui.csv', 'w', encoding='utf-8') as csv_file:
 writer = csv.writer(csv_file)
 header = ['decoded_title']
 writer.writerow(header)
 for decoded_title in data:
  writer.writerow([decoded_title])
```
